You are collaborating with a research group who are trying to see the affect of exercise on mental health. As part of this, they are using fitbits to monitor heart rate and want you to help them determine when the participants are exercising, walking, or resting. Implement a GaussianHMM from hmmlearn (or yur choice of another package) to help them in this process

A. Using the data and no prior knowledge, fit a Gaussian HMM and report the means and covariances of each state. Using the means, infer which state is resting, which is walking, and which is exercising (5 points)

```python
import pandas as pd
import numpy as np
from hmmlearn import hmm
from hmmlearn.hmm import GaussianHMM
data = pd.read_csv('heart_rates.csv')
data
```

```python
observations = data["Heart Rate"]
```

```python
print(observations)
```

```python
#create a GaussianHMM model object with 3 components
hmm = GaussianHMM(n_components=3, covariance_type="full", n_iter=100, random_state=54)
```

```python
#now fit in on the data
hmm.fit(data)
```

```python
#extract the means and covariances
means=hmm.means_
covariances = hmm.covars_
```

```python
print(means)
```

```python
print(covariances)
```

```python
#now predict the states
predicted_states = hmm.predict(data)
```

```python
#assign values to each state level
states = {0:"Rest", 1:"Walk", 2:"Exercise"}
```

```python
#call them what they are
predicted_state_names = [states[state] for state in predicted_states]
```

```python
#get the data classified
classified_data = np.hstack((data, predicted_states.reshape(-1,1)))
```
